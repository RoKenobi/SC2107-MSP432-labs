```python
# Import required libraries
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the model: Distance = A / ADC - B
def model(adc, A, B):
    return A / adc - B

# === ENTER YOUR DATA HERE ===
# Format: [distance_in_mm], [adc_values]
distances = np.array([100, 150, 200, 250])        # in mm
adc_values = np.array([11854, 7749, 6826, 5413])  # your nc values (center sensor)

# Optional: Add more points later, e.g. [300, 400, 500]

# Perform curve fitting
popt, pcov = curve_fit(model, adc_values, distances, p0=[1_400_000, 20])
A_fit, B_fit = popt

print(f"Best-fit parameters:")
print(f"A = {A_fit:.0f}")
print(f"B = {B_fit:.2f}")

# Plot the result
adc_fine = np.linspace(adc_values.min(), adc_values.max(), 100)
dist_fine = model(adc_fine, A_fit, B_fit)

plt.figure(figsize=(8,5))
plt.scatter(adc_values, distances, color='red', label='Measured Data')
plt.plot(adc_fine, dist_fine, 'b--', label=f'Fit: Distance = {A_fit:.0f}/ADC - {B_fit:.1f}')
plt.xlabel('Filtered ADC Value (nc)')
plt.ylabel('Distance (mm)')
plt.title('IR Sensor Calibration  Center Sensor')
plt.legend()
plt.grid(True)
plt.gca().invert_xaxis()  # ADC decreases as distance increases
plt.show()

# Validation with 10 mm tolerance
print("Validation (10 mm tolerance):")
tolerance = 10
for d_true, adc in zip(distances, adc_values):
    d_pred = model(adc, A_fit, B_fit)
    error = d_pred - d_true
    if abs(error) <= tolerance:
        status = ""  # Green tick (you can copy-paste this)
    else:
        status = ""  # Red cross
    print(f"ADC={adc:5d}  True={d_true:3d} mm, Predicted={d_pred:6.1f} mm ({error:+5.1f} mm) {status}")

print(f"\n  Use this in C code:  distance = ({int(A_fit)} / nc) - {int(round(B_fit))};\n")    
```
